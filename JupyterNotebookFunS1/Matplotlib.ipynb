## Goals of Data Vis
1. Clearly and accurately represent data
1. Be creative, with the goal of increasing readability
1. Label units, axes, and points of interests

## Some Jargon
Chart: 2D visualization
* Plot: chart of data points (e.g. scatter plot)
* Graph: chart of a mathematical function (e.g. sine curve)

## Ways to use Matplotlib
1. Using the pyplot module (e.g. like a state machine where there is a "current" figure)
1. Using the OOP interface (e.g. maintain object references)
1. Mix of the two

## Line Charts

```python
import matplotlib.pyplot as plt 

def line_chart_example(x, y, y3):
    plt.figure() # creates a new "current" figure
    plt.plot(x, y, c="green", lw=5, label="$y=x^2$")
    # plt.figure()
    plt.plot(x, y3, label="$y=x^3$")
    # add some labels
    plt.legend()
    plt.title("Our First Line Chart!!")
    plt.xlabel("X Units")
    plt.ylabel("Y Units")
    plt.grid(True)

    # customize the xticks
    xtick_labels = ["zero", "one", "two", "three this is a really long label", "four"]
    plt.xticks(x, xtick_labels, rotation=15, ha="right") # parallel lists

    plt.show()

# we need some x, y data
x = list(range(5))
y = [val ** 2 for val in x]
# task: add another line to the figure for y = x^3
y3 = [val ** 3 for val in x]
line_chart_example(x, y, y3)
```
